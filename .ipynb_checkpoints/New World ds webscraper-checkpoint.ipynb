```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time 
import pandas as pd
```

```python
# all quests
with open('new_world_xp.csv', 'a') as file:
    file.write("quest_name; level; xp \n")
    
path = 'C:/Program Files (x86)\chromedriver.exe'
driver = webdriver.Chrome(path) 
driver.get("https://nwdb.info/db/quests/page/1")
time.sleep(1)

#closing the popup ad
close_button = driver.find_element_by_id("video-close-button")
close_button.click()
time.sleep(1)

# scraping quest data
for k in range(116):
# scraping quest data
    quest_name = driver.find_elements_by_xpath('//td[@class="ellipsis svelte-1h0aoe"]')
    lvl = driver.find_elements_by_xpath('//tbody[@class="align-middle"]/tr/td[3]')
    xp = driver.find_elements_by_xpath('//tbody[@class="align-middle"]/tr/td[4]')

    with open('new_world_xp.csv', 'a') as file:
        for i in range(len(quest_name)):
            file.write(quest_name[i].text + ";" + 
                       lvl[i].text + ";" + 
                       xp[i].text + "\n")

    #click next page button
    next = driver.find_element_by_xpath('//ul[@class="ul-pagination pagination svelte-1h0aoe"]/li[6]/a')
    next.click()
     
    time.sleep(10)
    
driver.quit()
```
